```python
# Coordenadas del área a recortar (xmin, ymin, xmax, ymax)
x_min = int(coords[0])
y_min = int(coords[1])
x_max = int(coords[2])
y_max = int(coords[3])

output_video_path = f'/content/drive/MyDrive/ColabNotebooks/test/test{rand}/cut.mp4'  # Ruta del video de salida
# Abre el video
cap = cv2.VideoCapture(input_video_path)

# Verifica si el video se abrió correctamente
if not cap.isOpened():
    print("Error al abrir el video")
    exit()

# Obtén el ancho y alto del video
cap = cv2.VideoCapture(input_video_path)

# Verifica si el video se abrió correctamente
if not cap.isOpened():
    print("Error al abrir el video")
    exit()

# Obtén el ancho y alto del video
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))


# Aquí, 30 es el framerate del video original. Ajusta esto si es necesario.
out = cv2.VideoWriter(output_video_path, cv2.VideoWriter_fourcc(*'mp4v'), 30, (x_max-x_min, y_max-y_min))

while cap.isOpened():
    ret, frame = cap.read()

    if ret:
        # Recorta el área del fotograma
        crop_frame = frame[y_min:y_max, x_min:x_max]

        # Escribe el fotograma recortado en el nuevo video
        out.write(crop_frame)

        # Presiona 'q' para salir
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break

# Libera los objetos VideoCapture y VideoWriter
cap.release()
out.release()
cv2.destroyAllWindows()

print(f"Video recortado exportado a: {output_video_path}")
```

## Conteo de ropa desalarmada en video.

```python


def calculate_inner(box1, box2):
    #NOTE: box1 y box2 son listas de coordenadas [xmin, ymin, xmax, ymax]
    x1 = max(box1[0], box2[0])
    y1 = max(box1[1], box2[1])
    x2 = min(box1[2], box2[2])
    y2 = min(box1[3], box2[3])

    #NOTE: Calcula el área de intersección
    inter_area = max(0, x2 - x1) * max(0, y2 - y1)

    #NOTE: Calcula el área de ambas cajas
    box1_area = (box1[2] - box1[0]) * (box1[3] - box1[1])
    box2_area = (box2[2] - box2[0]) * (box2[3] - box2[1])

    #NOTE: Calcula el área de unión
    union_area = box1_area + box2_area - inter_area

    #NOTE: Calcula IoU
    inner = inter_area / union_area
    return inner

#NOTE: Función para verificar si un marco está dentro de otro
def is_within(inner_box, outer_box):
    inner_xmin, inner_ymin, inner_xmax, inner_ymax = inner_box
    outer_xmin, outer_ymin, outer_xmax, outer_ymax = outer_box
    return (inner_xmin > outer_xmin and inner_ymin > outer_ymin and
            inner_xmax < outer_xmax and inner_ymax < outer_ymax)

n = 1;
x=0
ropa = ""

#NOTE: Trabajar las caracteristicas obtenidas por YOLO
for result in results:
    boxes = result.boxes  # Boxes object for bounding box outputs
    masks = result.masks  # Masks object for segmentation masks outputs
    keypoints = result.keypoints  # Keypoints object for pose outputs
    probs = result.probs  # Probs object for classification outputs
    obb = result.obb  # Oriented boxes object for OBB outputs
    clases = boxes.cls.cpu().numpy()
    conf = boxes.conf.cpu().numpy()
    if 0 in clases:
      if len(clases) > 2:
        inner = calculate_inner(boxes.xyxy.cpu().numpy()[1], boxes.xyxy.cpu().numpy()[2]) #<-- saber si la alarma se encuentra sobre el iman en caso de que su frame se encuentre parcialmente y no totalemente dentro del otro
        trash = is_within(boxes.xyxy.cpu().numpy()[2],boxes.xyxy.cpu().numpy()[0])   #<-- saber si la alarma estasobre el basurero
        iman =  is_within(boxes.xyxy.cpu().numpy()[2], boxes.xyxy.cpu().numpy()[1])  #<-- saber si la alarama esta sobre el iman
        conf2 = conf[2]
        if trash and n == 0 and conf[0] > 0.35:                #<-- contabilizar cuantas veces tiraron una alarma al basurero
          result.save(filename=f"te/basurero{x}.jpg")
          n = 1
          ropa += "t"
          x += 1
        elif (inner > 0.1 or iman) and conf2 > 0.60:            #<-- contabilizar en cuantos frames la alarma estubo sobre el iman
          result.save(filename=f"te/iman{x}.jpg")
          n = 0
          ropa += "m"
          x += 1
      else:
        if 1 in clases:
          conf2 = conf[1]
          inner = calculate_inner(boxes.xyxy.cpu().numpy()[1], boxes.xyxy.cpu().numpy()[0])
          iman =  is_within(boxes.xyxy.cpu().numpy()[1], boxes.xyxy.cpu().numpy()[0])
          if (inner > 0.1 or iman) and conf2 > 0.6:            #<-- contabilizar en cuantos frames la alarma estubo sobre el iman
            result.save(filename=f"te/iman{x}.jpg")
            n = 0
            ropa += "m"
            x += 1
        elif 2 in clases:
          trash = is_within(boxes.xyxy.cpu().numpy()[1],boxes.xyxy.cpu().numpy()[0])
          if trash and n == 0 and conf[0] > 0.35:                #<-- contabilizar cuantas veces tiraron una alarma al basurero
            result.save(filename=f"te/basurero{x}.jpg")
            n = 1
            ropa += "t"
            x += 1
    elif (1 in clases and n == 0 and not 0 in clases and not 2 in clases):
      result.save(filename=f"te/sin_alarma{x}.jpg")
      n = 1
      ropa += "t"
      x += 1




#NOTE: Resultados
salida = ropa.split("t")    #<-- buffer salida de bucle.
resultado =  [subarray for subarray in salida if subarray] #<-- eliminacion de basura en buffer.
print(resultado)
print(len(resultado))       #<-- impresion de resultado.
```
